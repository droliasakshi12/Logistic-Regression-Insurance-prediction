#  üß¨Insurance Logistic Regression 

## üîóImporting Required library 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

## üìä Loading insurance_data dataset 

##### Using pandas library to :
- Read the csv file 
- Data Filtering
- Data Cleaning  


```python
#loading the data set using read_csv 
#used the backslash as it it inside the spam_ham_dataset.csv folder 
df = pd.read_csv("insurance_data.csv")
```

```python
#printing the data 
df
```

## üî¢ Checking Null Values 

```python
df.isnull().sum()
```

## üìàData Visualisation

```python
plt.scatter(df.age,df.bought_insurance,color="blue")
plt.xlabel("Age")
plt.ylabel("Bought Insurance")
plt.show()
```

## üßë‚ÄçüîßTraning and Testing Data 

```python
#using sklearn to import ttrain test and fitting to the model 
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

## ü§ñLoading the Model 

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#fitting the data to the model 
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
#checking the performance of the model by using score function
model.score(x_test,y_test)
```
